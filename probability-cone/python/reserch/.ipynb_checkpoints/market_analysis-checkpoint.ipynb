```python
def make_n_shifts(data, col_name, n):
    for i in  range(1, n+1):
        data[str(col_name + " -" + str(i))] = data[col_name].shift(-i)
```

```python
make_n_shifts(msft, "Open", 5)
msft
```

```python
data = msft[["Open", "Open -1", "Open -2", "Open -3", "Open -4", "Open -5"]].iloc[:-5,]
data
```

```python
scaler = MinMaxScaler()

X = scaler.fit_transform(data.iloc[:,:5].to_numpy())
X = sm.add_constant(X)
y = scaler.fit_transform(data.iloc[:,5:6].to_numpy())
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

model = sm.OLS(y_train, X_train).fit()
predictions = model.predict(X_test) 

print_model = model.summary()
print(print_model)
```

```python
plt.plot(y_test[::10], label="real values")
plt.plot(predictions[::10], label="predictions")
plt.ylabel("Price")
plt.xlabel("Iteration")
plt.legend();
```
